```python
import pandas as pd

# -----------------------------
# SAMPLE DATA
# -----------------------------

patients = pd.DataFrame({
    'patient_id': [1, 2, 3],
    'name': ['Alice Smith', 'John Doe', 'Maria Lee'],
    'age': [34, 52, 41],
    'gender': ['F', 'M', 'F']
})

appointments = pd.DataFrame({
    'appointment_id': [101, 102, 103, 104],
    'patient_id': [1, 2, 1, 3],
    'appointment_date': ['2024-01-10', '2024-01-12', '2024-01-15', '2024-01-20'],
    'status': ['Completed', 'Cancelled', 'Completed', 'No-Show']
})

diagnoses = pd.DataFrame({
    'diagnosis_id': [201, 202, 203],
    'patient_id': [1, 2, 3],
    'diagnosis': ['Diabetes', 'Hypertension', 'Asthma']
})

patients, appointments, diagnoses
```

```python
sql_practice = {
    "Q1": {
        "question": "Select all female patients.",
        "answer": """
SELECT *
FROM Patients
WHERE gender = 'F';
"""
    },
    "Q2": {
        "question": "Count how many appointments were completed.",
        "answer": """
SELECT COUNT(*) AS completed_count
FROM Appointments
WHERE status = 'Completed';
"""
    },
    "Q3": {
        "question": "List all appointments with the patient’s name.",
        "answer": """
SELECT 
    Patients.name AS PatientName,
    Appointments.appointment_date AS AppointmentDate,
    Appointments.status AS Status
FROM Appointments
JOIN Patients 
    ON Appointments.patient_id = Patients.patient_id;
"""
    },
    "Q4": {
        "question": "Find patients who have never had an appointment.",
        "answer": """
SELECT Patients.name
FROM Patients
LEFT JOIN Appointments 
    ON Patients.patient_id = Appointments.patient_id
WHERE Appointments.patient_id IS NULL;
"""
    },
    "Q5": {
        "question": "Count appointments per patient.",
        "answer": """
SELECT 
    Patients.name,
    COUNT(Appointments.appointment_id) AS appointment_count
FROM Patients
LEFT JOIN Appointments 
    ON Patients.patient_id = Appointments.patient_id
GROUP BY Patients.name
ORDER BY appointment_count DESC;
"""
    },
    "Q6": {
        "question": "Show all diagnoses for patients over age 40.",
        "answer": """
SELECT 
    Patients.name,
    Patients.age,
    Diagnoses.diagnosis
FROM Patients
JOIN Diagnoses 
    ON Patients.patient_id = Diagnoses.patient_id
WHERE Patients.age > 40;
"""
    },
    "Q7": {
        "question": "Find the most common appointment status.",
        "answer": """
SELECT status, COUNT(*) AS count_status
FROM Appointments
GROUP BY status
ORDER BY count_status DESC
LIMIT 1;
"""
    },
    "Q8": {
        "question": "Return patients who have both a diagnosis AND at least one completed appointment.",
        "answer": """
SELECT DISTINCT Patients.name
FROM Patients
JOIN Diagnoses 
    ON Patients.patient_id = Diagnoses.patient_id
JOIN Appointments 
    ON Patients.patient_id = Appointments.patient_id
WHERE Appointments.status = 'Completed';
"""
    },
    "Q9": {
        "question": "Show all appointments after 2024‑01‑12.",
        "answer": """
SELECT appointment_id, patient_id, appointment_date
FROM Appointments
WHERE appointment_date > '2024-01-12';
"""
    },
    "Q10": {
        "question": "Label patients as Senior (>=50) or Adult (<50).",
        "answer": """
SELECT 
    name,
    age,
    CASE 
        WHEN age >= 50 THEN 'Senior'
        ELSE 'Adult'
    END AS age_group
FROM Patients;
"""
    }
}

sql_practice
```
