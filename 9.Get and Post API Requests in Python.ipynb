```python
import requests

# ------------------------------------------------------------
# Define the payload (the data you want to send as query params)
# ------------------------------------------------------------
payload = {
    "firstName": "John",
    "lastName": "Doe"
}

# ------------------------------------------------------------
# Send a GET request to httpbin.org
# httpbin.org/get simply returns whatever you send to it.
# Using 'params=payload' attaches the data to the URL as query parameters.
# ------------------------------------------------------------
r = requests.get('https://httpbin.org/get', params=payload)

# ------------------------------------------------------------
# Print the HTTP status code to confirm the request succeeded.
# 200 = OK
# ------------------------------------------------------------
print(r.status_code)

# ------------------------------------------------------------
# Print the full response text so you can see how the server
# received and returned your parameters.
# ------------------------------------------------------------
print(r.text)
```

```python
import requests

# ------------------------------------------------------------
# Define the payload you want to send in the POST request.
# Using 'data=' sends this as FORM data (application/x-www-form-urlencoded).
# ------------------------------------------------------------
payload = {
    "firstName": "John",
    "lastName": "Doe"
}

# ------------------------------------------------------------
# Send a POST request to httpbin.org/post.
# httpbin.org is a testing service that simply returns whatever
# you send to it, so you can inspect how your request was handled.
# ------------------------------------------------------------
r = requests.post('https://httpbin.org/post', data=payload)

# ------------------------------------------------------------
# Print the HTTP status code.
# 200 means the request was successful.
# ------------------------------------------------------------
print(r.status_code)

# ------------------------------------------------------------
# Print the full response body as text.
# This shows you exactly how the server interpreted your request,
# including the form data you sent.
# ------------------------------------------------------------
print(r.text)
```
