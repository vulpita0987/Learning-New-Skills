```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


dataset_as_given = pd.read_csv(r"d:\Desktop\Practice Python\21.gym_members_exercise_tracking.csv")
print(dataset_as_given)

# Get column names
column_names = dataset_as_given.columns
print(column_names)

# Using describe() to calculate summary statistics
basic_stats = dataset_as_given.describe()
print(basic_stats)

# Using describe() to calculate summary statistics
basic_stats = dataset_as_given.describe()
print(basic_stats)

desired = ["Age", "Weight (kg)", "Height (m)", "Max_BPM", "Avg_BPM", "Resting_BPM", 
           "Session_Duration (hours)", "Calories_Burned", "Fat_Percentage", "Water_Intake (liters)", "Workout_Frequency (days/week)", "Experience_Level", "BMI"]

dataset_as_given_numeric = dataset_as_given[desired]
# Calculating mean, median, and mode
mean_values = dataset_as_given_numeric.mean()
median_values = dataset_as_given_numeric.median()
mode_values = dataset_as_given_numeric.mode().iloc[0] 

print("Means for numerical columns in dataset:")
print(mean_values)
print("Medians for numerical columns in dataset:")
print(median_values)
print("Modes for numerical columns in dataset:")
print(mode_values)


# Calculating correlation between columns
correlation_numerical_columns = dataset_as_given_numeric.corr()
print(correlation_numerical_columns)


#Correlation matrix - visual
matrix = correlation_numerical_columns.corr()

plt.figure(figsize=(8,6))
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()


# Calculating variance and standard deviation
variance_val = correlation_numerical_columns.var()
std_dev_val = correlation_numerical_columns.std()

print("Variance for numerical columns in dataset:")
print(variance_val)
print("Standard Deviation for numerical columns in dataset:")
print(std_dev_val)

# Calculating skewness and kurtosis
skewness_val = correlation_numerical_columns.skew()
kurtosis_val = correlation_numerical_columns.kurt()

print("Skewness for numerical columns in dataset:")
print(skewness_val)
print("Kurtosis for numerical columns in dataset:")
print(kurtosis_val)


# Data preparation
penguins_grouped = dataset_as_given[['Age', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Age'], color='skyblue')

plt.title('Average Age by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Age')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Age', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Age'], color='skyblue')

plt.title('Average Age by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Age')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

############################################################


# Data preparation
penguins_grouped = dataset_as_given[['Weight (kg)', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Weight (kg)'], color='skyblue')

plt.title('Average Weight (kg) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Weight (kg)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Weight (kg)', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Weight (kg)'], color='skyblue')

plt.title('Average Weight (kg) by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Weight (kg)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

######################################################################################


# Data preparation
penguins_grouped = dataset_as_given[['Height (m)', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Height (m)'], color='skyblue')

plt.title('Average Height (m) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Height (m)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Height (m)', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Height (m)'], color='skyblue')

plt.title('Average Height (m) by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Height (m)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
################################################################


# Data preparation
penguins_grouped = dataset_as_given[['Max_BPM', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Max_BPM'], color='skyblue')

plt.title('Average Max_BPM by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Max_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Max_BPM', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Max_BPM'], color='skyblue')

plt.title('Average Max_BPM by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Max_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
########################################################################



# Data preparation
penguins_grouped = dataset_as_given[['Avg_BPM', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Avg_BPM'], color='skyblue')

plt.title('Average Avg_BPM by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Avg_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Avg_BPM', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Avg_BPM'], color='skyblue')

plt.title('Average Avg_BPM by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Avg_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

#################################################################


     

# Data preparation
penguins_grouped = dataset_as_given[['Resting_BPM', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Resting_BPM'], color='skyblue')

plt.title('Average Resting_BPM by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Resting_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Resting_BPM', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Resting_BPM'], color='skyblue')

plt.title('Average Resting_BPM by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Resting_BPM')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

##########################################################


# Data preparation
penguins_grouped = dataset_as_given[['Session_Duration (hours)', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Session_Duration (hours)'], color='skyblue')

plt.title('Average Session_Duration (hours) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Session_Duration (hours)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Session_Duration (hours)', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Session_Duration (hours)'], color='skyblue')

plt.title('Average Session_Duration (hours) by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Session_Duration (hours)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
##################################################################


# Data preparation
penguins_grouped = dataset_as_given[['Calories_Burned', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Calories_Burned'], color='skyblue')

plt.title('Average Calories_Burned by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Calories_Burned')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Calories_Burned', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Calories_Burned'], color='skyblue')

plt.title('Average Calories_Burned by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Calories_Burned')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

#####################################################


# Data preparation
penguins_grouped = dataset_as_given[['Fat_Percentage', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Fat_Percentage'], color='skyblue')

plt.title('Average Fat_Percentage by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Fat_Percentage')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Fat_Percentage', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Fat_Percentage'], color='skyblue')

plt.title('Average Fat_Percentage by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Fat_Percentage')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

############################################################################


# Data preparation
penguins_grouped = dataset_as_given[['Water_Intake (liters)', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Water_Intake (liters)'], color='skyblue')

plt.title('Average Water_Intake (liters) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Water_Intake (liters)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Water_Intake (liters)', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Water_Intake (liters)'], color='skyblue')

plt.title('Average Water_Intake (liters) by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Water_Intake (liters)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()

#########################################################################

      

# Data preparation
penguins_grouped = dataset_as_given[['Workout_Frequency (days/week)', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Workout_Frequency (days/week)'], color='skyblue')

plt.title('Average Workout_Frequency (days/week) by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Workout_Frequency (days/week)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Workout_Frequency (days/week)', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Workout_Frequency (days/week)'], color='skyblue')

plt.title('Average Workout_Frequency (days/week) by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Workout_Frequency (days/week)')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
###################################################################

      
 

# Data preparation
penguins_grouped = dataset_as_given[['Experience_Level', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['Experience_Level'], color='skyblue')

plt.title('Average Experience_Level by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Experience_Level')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['Experience_Level', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['Experience_Level'], color='skyblue')

plt.title('Average Experience_Level by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average Experience_Level')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()
###########################################################################

      
 

# Data preparation
penguins_grouped = dataset_as_given[['BMI', 'Gender']].groupby('Gender').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Gender'], penguins_grouped['BMI'], color='skyblue')

plt.title('Average BMI by Gender')
plt.xlabel('Gender')
plt.ylabel('Average BMI')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()


#############################################################################################################
# Data preparation
penguins_grouped = dataset_as_given[['BMI', 'Workout_Type']].groupby('Workout_Type').mean().reset_index()

# Creating a bar chart
plt.figure(figsize=(6,4))
bars = plt.bar(penguins_grouped['Workout_Type'], penguins_grouped['BMI'], color='skyblue')

plt.title('Average BMI by Workout_Type')
plt.xlabel('Workout_Type')
plt.ylabel('Average BMI')

# Add value labels on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # x position (center of bar)
        height,                            # y position (top of bar)
        f'{height:.1f}',                   # text (formatted to 1 decimal)
        ha='center', va='bottom'           # alignment
    )

plt.tight_layout()
plt.show()



```
