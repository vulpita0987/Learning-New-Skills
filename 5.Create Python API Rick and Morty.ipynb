```python
import requests
import pandas as pd

# ------------------------------------------------------------
# FUNCTION: main_request
# ------------------------------------------------------------
# Sends a GET request to the Rick & Morty API for a specific page.
# The API uses pagination, so we must pass ?page=X to retrieve each page.
# The f-string ensures the page number is correctly inserted.
# ------------------------------------------------------------
def main_request(baseurl, endpoint, x):
    r = requests.get(f"{baseurl}{endpoint}?page={x}")
    return r.json()  # Convert JSON response into a Python dictionary


# ------------------------------------------------------------
# FUNCTION: get_pages
# ------------------------------------------------------------
# The API response contains an "info" section with metadata.
# "pages" tells us how many pages of results exist.
# We use this to loop through all pages later.
# ------------------------------------------------------------
def get_pages(response):
    pages = response['info']['pages']
    return pages


# ------------------------------------------------------------
# FUNCTION: parse_json
# ------------------------------------------------------------
# Extracts the fields we care about from a single page of results.
# Each page contains up to 20 characters.
#
# For each character, we extract:
#   - id: unique character ID
#   - name: character name
#   - no_ep: number of episodes they appear in
#
# We store each character as a dictionary and return a list of them.
# ------------------------------------------------------------
def parse_json(response):
    charlist = []

    for item in response['results']:
        char = {
            'id': item['id'],                     # Character ID
            'name': item['name'],                 # Character name
            'no_ep': len(item['episode']),        # Count of episodes
        }
        charlist.append(char)

    return charlist


# ------------------------------------------------------------
# MAIN EXECUTION
# ------------------------------------------------------------

baseurl = "https://rickandmortyapi.com/api/"
endpoint = "character"

mainlist = []  # This will store ALL characters from ALL pages

# First request to determine how many pages exist
data = main_request(baseurl, endpoint, 1)
total_pages = get_pages(data)

print("Total pages:", total_pages)
print("Sample from first page:", parse_json(data))


# ------------------------------------------------------------
# Loop through ALL pages and collect ALL characters
# ------------------------------------------------------------
for x in range(1, total_pages + 1):
    print(f"Fetching page {x}...")
    page_data = main_request(baseurl, endpoint, x)
    mainlist.extend(parse_json(page_data))  # Add parsed characters to main list


print("Total characters collected:", len(mainlist))


# ------------------------------------------------------------
# Convert the final list into a DataFrame
# ------------------------------------------------------------
df = pd.DataFrame(mainlist)

print("\nFirst 5 rows:")
print(df.head())

print("\nLast 5 rows:")
print(df.tail())


# ------------------------------------------------------------
# Save to CSV
# ------------------------------------------------------------
df.to_csv('charlist.csv', index=False)
print("\nSaved to charlist.csv")
```
