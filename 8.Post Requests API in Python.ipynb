```python
import requests
```

```python
#401 = “You’re not authorized to access this endpoint.”
r = requests.get('https://api.github.com/user', auth = ('user', 'pass'))
print(r.status_code)
```

```python
res = requests.post('https://httpbin.org/post',
                    json = {"key": "value"})
print(res.text)
```

```python

import json

data = {'user': 'me@example.com'  }
res = requests.post('https://httpbin.org/post',
                    data = json.dumps(data))
print(res.text)
```
