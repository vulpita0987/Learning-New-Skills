REST API </br>
REST = Representational State Transfer -> communicating over the web (interent)</br>
API = Application Programming Interface (a way two pieces of software can communicate with one another)</br>
You will be getting JSON data</br>
JSON - JavaScript Object Notation</br>
End Points -> spcific URL where an API receives requests and sends responses</br>
Exposed API endpoints can be used to retrive data through and API from a Server</br>  
APIs are used to get two different applications to talk -> connecting directly to the database would create security concerns</br>
The backend can change - such as the programming language used for an application can change all together - and the API can continue having the exposed endpoint - allowing the part of the application that the user sees AKA the front end to not be affected due to the API use - that extracts the required data regardless of the programming language</br>
Allows for multiple people to access data and create new apps from the data - allowing the data to be consumed</br>
GET is used to retrive data from the server</br>
POST is used to write data</br>
DELETE is used to delete data</br>
PUT is used to update data</br>
CRUD operations -> Create, Read, Update, Delete</br> 
PATCH can be used to UPDATE sections of a data entry</br>
