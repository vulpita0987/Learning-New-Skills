```python
import requests
import json

url = "https://httpbin.org/post"

data = {
'name': 'ScrapeOps',
'title': 'Python Webscraping Playbook',
'url': 'scrapeops.io/python-webscraping-playbook'

}

#Convert data to JSON formatting
json_data = json.dumps(data)

headers = {'Content-Type': 'application/json'}
#sent POST request with FORM data using the data parameter
response = requests.post(url, data=data, headers=headers)

print(response.text)
```

```python
import requests
import json

url = "https://httpbin.org/post"

data = {
'name': 'ScrapeOps',
'title': 'Python Webscraping Playbook',
'url': 'scrapeops.io/python-webscraping-playbook'

}

#Create a session object
session = requests.Session()

#Set the Content-Type header to application/json for all requests in the session
session.headers.update({'content-Type': 'application/json'})

#Send a POST request with JSON data using the session object
response = session.post(url, json=data)

#Print the response
print(response.text)
```
